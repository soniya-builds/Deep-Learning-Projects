```python
import numpy as np
data_array = np.load("data_array.npy")
label_array = np.load("label_array.npy")
group_array = np.load("group_array.npy")
```

```python
epochs_array = np.load("data_array.npy")
epochs_labels = np.load("label_array.npy")
groups_array = np.load("group_array.npy")

print(epochs_array.shape)
```

```python
from tensorflow.keras.layers import Conv1D, BatchNormalization, LeakyReLU, MaxPool1D, \
GlobalAveragePooling1D, Dense, Dropout, AveragePooling1D

from tensorflow.keras.models import Sequential

from tensorflow.keras.backend import clear_session
```

```python
epochs_array = np.moveaxis(data_array, 1, 2)
epochs_labels = label_array

print("CNN Input shape:", epochs_array.shape)
```

```python
def cnnmodel():
    clear_session()
    
    model = Sequential()
    
    model.add(Conv1D(filters=5, kernel_size=3, strides=1, input_shape=(1250, 19)))
    model.add(BatchNormalization())
    model.add(LeakyReLU())
    model.add(MaxPool1D(pool_size=2, strides=2))
    
    model.add(Conv1D(filters=5, kernel_size=3, strides=1))
    model.add(LeakyReLU())
    model.add(MaxPool1D(pool_size=2, strides=2))
    model.add(Dropout(0.5))
    
    model.add(Conv1D(filters=5, kernel_size=3, strides=1))
    model.add(LeakyReLU())
    model.add(AveragePooling1D(pool_size=2, strides=2))
    model.add(Dropout(0.5))
    
    model.add(Conv1D(filters=5, kernel_size=3, strides=1))
    model.add(LeakyReLU())
    model.add(AveragePooling1D(pool_size=2, strides=2))
    
    model.add(Conv1D(filters=5, kernel_size=3, strides=1))
    model.add(LeakyReLU())
    
    model.add(GlobalAveragePooling1D())
    
    model.add(Dense(1, activation='sigmoid'))
    
    model.compile(
        optimizer='adam',
        loss='binary_crossentropy',
        metrics=['accuracy']
    )
    
    return model


#create model
model = cnnmodel()
model.summary()

#Train Model
history = model.fit(
    epochs_array,
    epochs_labels,
    epochs=10,
    batch_size=32,
    validation_split=0.2
)
```
